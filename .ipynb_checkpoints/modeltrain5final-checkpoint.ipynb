## Project Overview: Training and Comparison of MobileNetV2 Models

In this project, we trained and evaluated two variants of the MobileNetV2 architecture for binary image classification:

- **Initial Model:**  
  Trained with the MobileNetV2 base (pre-trained on ImageNet) with all base layers frozen, allowing only the custom classification head to learn from our data.

- **Fine-tuned Model:**  
  Started from the same pre-trained MobileNetV2 base, but selectively unfroze and fine-tuned some of the deeper layers alongside the classification head to better adapt to our specific dataset.

### Why MobileNetV2 over ResNet?

We chose **MobileNetV2** as our base model instead of ResNet for the following reasons:
- **Efficiency:** MobileNetV2 is a lightweight, highly efficient model designed for fast inference and lower memory usage, making it ideal for deployment on resource-constrained devices.
- **Proven Performance:** Despite its smaller size, MobileNetV2 has demonstrated strong performance on a wide range of visual recognition tasks, often matching or surpassing larger models on practical datasets.
- **Speed:** Training and fine-tuning MobileNetV2 is faster compared to deeper models like ResNet, enabling quicker experimentation and iteration.

These advantages make MobileNetV2 a strong baseline for transfer learning in real-world scenarios, especially when computational resources are limited.


```python
import os

def count_images_in_classes(base_dir):
    for split in ['train', 'test']:
        split_path = os.path.join(base_dir, split)
        print(f"\n{split.upper()}:")
        for class_name in os.listdir(split_path):
            class_path = os.path.join(split_path, class_name)
            count = len([f for f in os.listdir(class_path) if f.lower().endswith(('.jpg','.jpeg','.png'))])
            print(f"  {class_name}: {count} images")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Train Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2  
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Generators
train_gen = train_datagen.flow_from_directory(
    'resnet_dataset/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    'resnet_dataset/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

test_gen = test_datagen.flow_from_directory(
    'resnet_dataset/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Load pre-trained MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze base layers

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)

model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# Callbacks
checkpoint = ModelCheckpoint('best_mobilenetv2.h5', monitor='val_accuracy', save_best_only=True, mode='max')
early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)

# Train
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=30,
    callbacks=[checkpoint, early_stop]
)
# Save the complete trained model in Keras native format
model.save('mobilenetv2_custom_final.keras')
print("Model saved as mobilenetv2_custom_final.keras")
```

```python
# For example, using a test generator named test_gen
test_loss, test_acc = model.evaluate(test_gen)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Evaluate on validation set
val_loss, val_acc = model.evaluate(val_gen)
print(f"Validation Accuracy after initial training: {val_acc * 100:.2f}%")
print(f"Validation Loss: {val_loss:.4f}")

# Evaluate on test set
test_loss, test_acc = model.evaluate(test_gen)
print(f"Test Accuracy after initial training: {test_acc * 100:.2f}%")
print(f"Test Loss: {test_loss:.4f}")
```

```python
# Unfreeze last 20 layers of MobileNetV2
for layer in base_model.layers[-20:]:
    layer.trainable = True

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Fine-tune
history_ft = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    callbacks=[checkpoint, early_stop]
)
```

```python
# Save the initial trained model
model.save('mobilenetv2_initial_model.keras')
print("Initial MobileNetV2 model saved as mobilenetv2_initial_model.keras")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# 1. Load the best initial model
model = load_model('mobilenetv2_initial_model.keras')
print("Loaded initial MobileNetV2 model.")

# 2. Unfreeze last 10 layers of base model
for layer in model.layers[-10:]:
    if not isinstance(layer, tf.keras.layers.BatchNormalization):
        layer.trainable = True

# 3. Compile with a smaller learning rate
model.compile(optimizer=Adam(learning_rate=1e-6),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 4. Fine-tuning callbacks
checkpoint_ft = ModelCheckpoint('mobilenetv2_finetuned_best.h5', monitor='val_accuracy', save_best_only=True, mode='max')
early_stop_ft = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# 5. Fine-tune
history_ft = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=15,
    callbacks=[checkpoint_ft, early_stop_ft]
)

# 6. Save the fine-tuned model
model.save('mobilenetv2_finetuned_model.keras')
print("Fine-tuned model saved as mobilenetv2_finetuned_model.keras")
```

```python
val_loss_ft, val_acc_ft = model.evaluate(val_gen)
print(f"Validation Accuracy after fine-tuning: {val_acc_ft * 100:.2f}%")
test_loss_ft, test_acc_ft = model.evaluate(test_gen)
print(f"Test Accuracy after fine-tuning: {test_acc_ft * 100:.2f}%")
```

## Model Comparison: Initial vs Fine-tuned MobileNetV2

We trained and evaluated two different versions of MobileNetV2 on our dataset:

- **Initial MobileNetV2:**  
  - *Test Accuracy (Initial MobileNetV2): **0.9072***
  - This model was trained with the base MobileNetV2 (ImageNet weights) and custom classification layers. The base layers were frozen during training, so only the custom head was trained.

- **Fine-tuned MobileNetV2:**  
  - *Test Accuracy (Fine-tuned MobileNetV2): **0.8763***
  - For this model, we fine-tuned (unfroze) some of the base MobileNetV2 layers and continued training with a lower learning rate. The goal was to adapt the pre-trained features more closely to our data.

### **Observation**
- The **Initial MobileNetV2** achieved a higher test accuracy (**0.9072**) compared to the **Fine-tuned MobileNetV2** (**0.8763**).
- This suggests that, for our dataset, keeping the pre-trained weights frozen performed slightly better, while fine-tuning may have led to some overfitting or did not sufficiently improve feature extraction for our specific task.
- Further hyperparameter tuning or data augmentation may help improve fine-tuned results in the future.
